```python
# First import the libraries cv2 and matplotlib
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read an image file using cv2
img = cv2.imread("leo.png")
# we can read an image with filteration inside function 
image1 = cv2.imread("leo.png" , 0) # Gray-Scale-Level 
image2 = cv2.imread("leo.png" , cv2.IMREAD_GRAYSCALE) # Gray-Scale-Level 
```

```python
# what if print image as object directly 
img
```

```python
# that's printing (number_of_pixels_in_rows , number_of_pixels_in_heights , 3 mean 3 channels in per i pixel)
img.shape
```

```python
# Show an image file using cv2 => imshow(window_name , image_object)
cv2.imshow('Image',img)
# wait for action to close window
cv2.waitKey(0)
# destroy window after this action
cv2.destroyAllWindows()
```

```python
# transform image to gray using numpy
copy_img = np.copy(img)
gray_img = cv2.cvtColor(copy_img,cv2.COLOR_RGB2GRAY)
cv2.imshow('GRAY IMAGE',gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Read an image file with transform it to gray using cv2 without using numpy
img = cv2.imread("leo.png",cv2.IMREAD_GRAYSCALE)
# cv2.imshow('GRAY IMAGE',img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# to customize window
cv2.namedWindow('image' , cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to save image
cv2.imwrite('denairo2.png',img)
```
